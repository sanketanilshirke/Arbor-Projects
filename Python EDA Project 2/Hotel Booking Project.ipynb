# Adding columns

```python
df["Arrival_Date"] = pd.to_datetime(
    df["arrival_date_day_of_month"].astype(str) + "-" +
    df["arrival_date_month"].astype(str) + "-" +
    df["arrival_date_year"].astype(str),
    format="%d-%B-%Y"
)
```

```python
df['total_num_people'] = df['adults']+df['children']+df['babies']
```

```python
df.drop(columns=["adults", "children","babies"], inplace=True)
```

```python
df.drop(df[df['total_num_people']== 0].index ,inplace=True)        # Remove records where 0 in total_num_people
```

# Data Visualization

```python
# 1)Question no. 1 Which Type Of hotel have More number of bookings [ City hotels or Resort Hotels ]
hotel_value_counts = df['hotel'].value_counts()
hotel_value_counts.plot(kind='pie',figsize=(11,9),fontsize=20, autopct='%1.2f%%',)
plt.ylabel('')
plt.title('Most Preffered  Hotel',fontsize = 10)
plt.show()
```

```python
# 2)Booking Count by Customer Type & Hotel

plt.figure(figsize=(10,6))

sns.countplot(data=df,x='customer_type',hue='hotel')

plt.title('Count of Bookings by Customer Type and Hotel')
plt.xlabel('Customer Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Hotel')

plt.show()
```

```python
# 3)What is the pecentage of Booking cancellation for both type
plt.figure(figsize=(8,8))

df['is_canceled'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    explode=(0, 0.08),
    labels=['Not Canceled', 'Canceled'],
    textprops={'fontsize':14}
)

plt.title('Overall Booking Cancellation Percentage', fontsize=16)
plt.ylabel('')   # remove default ylabel
plt.show()
```

```python
#4)What is the repeated percentage for customers?[ for both type of hotel ie. City hotels or Resort Hotels ]

is_repeated_guest = df['is_repeated_guest'].value_counts()

plt.figure(figsize=(8,8))
is_repeated_guest.plot(kind='pie',
    labels=['New Guest', 'Repeated Guest'],   # nicer labels
    autopct='%1.1f%%',
    startangle=90,
    explode=(0, 0.1),                          # highlight repeated guest
    shadow=True,
    textprops={'fontsize':14}
)
plt.ylabel('')
plt.title('Overall Repeated Guest Percentage', fontsize=16)
plt.show()
```

```python
# 5) What is optimal stay length in both types of hotel ie. City hotels and Resort Hotels

# Chart - 13 visualization code(optimal stay length)
stay_length =df.groupby(['hotel','total_stay']).agg('count').reset_index()
stay_length = stay_length.iloc[:, :3]
stay_length = stay_length.rename(columns={'is_canceled':'Number of stays'})
stay_length
```

```python
# 6)correlation heatmap
plt.figure(figsize=(11,9))
sns.barplot(x='total_stay',y='Number of stays', data=stay_length,hue='hotel')
plt.title('Optimal Stay Length in Both hotel types', fontsize=20)
plt.xlabel('total_stay_in_night in days',fontsize=15)
plt.ylabel('count of stays', fontsize=15)
plt.show()
```

```python
#7 correlation heatmap
# Correlation Heatmap visualization code
# Select only numeric columns
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(18,10))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    cbar_kws={'shrink':0.7}
)

plt.title('Correlation Heatmap of Numerical Columns', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```
